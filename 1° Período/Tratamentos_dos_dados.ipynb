```python
# carregando o DF
df = pd.read_excel(
    r"C:\Users\UNIVASSOURAS\Documents\Diego\script\postos-e-delegacias-da-pf.xlsx"
)

# atribui uma lista
list = ['REGIÃO NORTE', 'REGIÃO CENTRO-OESTE', 'REGIÃO SUDESTE', 'REGIÃO NORDESTE', 'REGIÃO SUL']

df = df.loc[~df['Superintendência Regional'].isin(list)]

# apresentando os 5 primeiros registros do DF
df.head()
```

```python
data_atual = None
for index, row in df.iterrows():
    if pd.notna(row['Superintendência Regional']):
        data_atual = row['Superintendência Regional']
    else:
        df.at[index, 'Superintendência Regional'] = data_atual

df.head()
```

```python
# exemplo de como criar uma função em Python
def apresenta(info):
    if info == 'Oi' or info == 'oi':
        return print(f"O que foi apresentado é {info}")
    else:
        return print(f'Informação diferente,o que foi apresentado é {info}')

apresenta("oi")
```

```python
def corrigir_nan(dataframe, col):
    data_atual = None
    for index, row in dataframe.iterrows():
        if pd.notna(row[col]):
            data_atual = row[col]
        else:
            dataframe.at[index, col] = data_atual
    
    return dataframe
```

```python
lista = df.columns.tolist()
for i in lista:
    df = corrigir_nan(df, i)

df.head()
```
